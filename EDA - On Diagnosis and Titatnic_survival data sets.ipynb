## titanicsurvival.csv

```python
import pandas as pd
import numpy as np
```

```python
df2=pd.read_csv('titanicsurvival.csv')
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
mode1=df2.Age.mode()
```

```python
mode1
```

```python
df2.Age.fillna(24.0,inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2['Sex']=df2['Sex'].map({'male':0,'female':1}).astype(int)
```

```python
m=df2.iloc[:,0:4].values
```

```python
m
```

```python
n=df2.iloc[:,4].values
```

```python
n
```

```python
from sklearn.model_selection import train_test_split
m_train,m_test,n_train,n_test=train_test_split(m,n,test_size=0.25,random_state=0)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
m_train=sc.fit_transform(m_train)
m_test=sc.transform(m_test)
```

```python
m_train
```

```python
m_test
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
```

```python
models=[]
models.append(('LR',LogisticRegression(solver='liblinear',multi_class='ovr')))
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('DT',DecisionTreeClassifier()))
models.append(('NB',GaussianNB()))
models.append(('SVM',SVC(gamma='auto')))
```

```python
import matplotlib.pyplot as pl
```

```python
results = []
names = []
res = []
for name, model in models:
    kfold = StratifiedKFold(n_splits=10, random_state=None)
    cv_results = cross_val_score(model, m_train, n_train, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    res.append(cv_results.mean())
    print('%s: %f' % (name, cv_results.mean()))

pl.ylim(.600, .999)
pl.bar(names, res, color ='maroon', width = 0.6)

pl.title('Algorithm Comparison')
pl.show()
```

```python
model=LogisticRegression(solver='liblinear',multi_class='ovr')
model.fit(m_train,n_train)
```

```python
n_pred=model.predict(m_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(n_test, n_pred)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of the Model: {0}%".format(accuracy_score(n_test, n_pred)*100))
```
